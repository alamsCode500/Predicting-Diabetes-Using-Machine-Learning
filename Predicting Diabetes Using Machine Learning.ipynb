## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.

try:
    # 1. Remove duplicate rows
    df.drop_duplicates(inplace=True)

    # 2. Handle unexpected or inconsistent values
    # Check unique values for 'gender'
    print("Unique values in 'gender':", df['gender'].unique())

    # Map unique values of 'gender'
    df['gender'] = df['gender'].map({'Male': 2, 'Female': 1, 'Other': 0}).fillna(-1)

    # Check unique values for 'smoking_history'
    print("Unique values in 'smoking_history':", df['smoking_history'].unique())

    # Map unique values of 'smoking_history'
    df['smoking_history'] = df['smoking_history'].map({'never': 5, 'No Info': 4, 'current': 3, 'former': 2, 'ever': 1, 'not current': 0}).fillna(-1)

    # 3. Handle missing values (this dataset there are none)
    df.fillna(df.mean(), inplace=True)

    # 4. Feature scaling (Standardizing numerical columns (age, bmi, HbA1c_level, blood_glucose_level)
    #from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    numerical_columns = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']
    df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

    # 5. Check final dataset after preprocessing
    print("\nAnalysis-Ready Dataset:")
    print(df.head())  # Print the first few rows to check the final format

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    print("\nDataset analysis preparation attempted.")
```

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

**Statement 1:**
"Individuals with hypertension have a higher mean blood glucose level than those without hypertension."

**Statement 2:**
"The BMI of individuals who are current smokers is significantly different from those who never smoked."

**Statement 3:** "The prevalence of diabetes is higher among individuals with heart disease compared to those without heart disease."

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**"Individuals with hypertension have a higher mean blood glucose level than those without hypertension."**

**State the Hypotheses:**

* **Null Hypothesis (𝐻0):**

There is no difference in the mean blood glucose levels of individuals with and without hypertension.

* **Alternate Hypothesis (𝐻𝑎):**

The mean blood glucose level of individuals with hypertension is higher than that of individuals without hypertension.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value

import scipy.stats as stats

try:
    # Separate blood glucose levels based on hypertension status
    hypertension_yes = df[df['hypertension'] == 1]['blood_glucose_level']
    hypertension_no = df[df['hypertension'] == 0]['blood_glucose_level']

    # Perform an independent t-test
    t_stat, p_value = stats.ttest_ind(hypertension_yes, hypertension_no, alternative='greater')

    # Print the t-statistic and p-value
    print(f"T-Statistic: {t_stat}")
    print(f"P-Value: {p_value}")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    print("\nStatistical test for Hypothetical Statement 1 attempted.")
```

##### Which statistical test have you done to obtain P-Value?

We performed an **Independent t-test** to obtain the p-value. This test compares the means of two independent groups to determine if there is a statistically significant difference between them.

##### Why did you choose the specific statistical test?

We choose the **Independent t-test** because:

* The hypothesis compares the mean blood glucose levels of two distinct and independent groups:

    1. Individuals with hypertension (hypertension = 1).
    2. Individuals without hypertension (hypertension = 0).  

* Blood glucose level is a **continuous variable**, and the t-test is appropriate for testing mean differences of such variables.

* The test is designed for situations where the two groups being compared are unrelated (independent).

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**"The BMI of individuals who are current smokers is significantly different from those who never smoked."**

**State the Hypotheses:**

* **Null Hypothesis (𝐻0):**

There is no significant difference in BMI between individuals who are current smokers and those who never smoked.

* **Alternate Hypothesis (𝐻𝑎):**

There is a significant difference in BMI between individuals who are current smokers and those who never smoked.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value

from scipy import stats

try:
    # Separate the data for current smokers and never smokers
    current_smokers_bmi = df[df['smoking_history'] == 3]['bmi']  # 'current' smokers
    never_smoked_bmi = df[df['smoking_history'] == 5]['bmi']  # 'never' smoked

    # Perform the Independent T-test
    t_stat, p_value = stats.ttest_ind(current_smokers_bmi, never_smoked_bmi)

    # Display the results
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    print("\nStatistical test performed.")
```

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

**"The prevalence of diabetes is higher among individuals with heart disease compared to those without heart disease."**

The prevalence of diabetes is not significantly different between individuals with heart disease and those without heart disease. (i.e., no difference in the diabetes prevalence)

* **Alternate Hypothesis (𝐻𝑎):**

The prevalence of diabetes is significantly higher among individuals with heart disease compared to those without heart disease. (i.e., higher prevalence in individuals with heart disease)

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value

import scipy.stats as stats

try:
    # Create a contingency table for 'heart_disease' and 'diabetes'
    contingency_table = pd.crosstab(df['heart_disease'], df['diabetes'])

    # Perform the Chi-Square Test of Independence
    chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

    # Output the p-value
    print(f"P-Value: {p_value}")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    print("\nStatistical test attempted.")
```

##### Why did you choose the specific statistical test?

* The **Chi-Square Test of Independence** is appropriate for testing the relationship between two categorical variables. In this case, both **heart_disease** and **diabetes** are binary (categorical) variables. ### 1. Handling Missing Values

```python
# Handling Missing Values & Missing Value Imputation

try:
    # Check for missing values in the dataset
    print("Missing values before handling:")
    print(df.isnull().sum())

    # Impute missing values for numerical columns with the mean
    numerical_columns = ['bmi', 'age', 'HbA1c_level', 'blood_glucose_level']
    for column in numerical_columns:
        if df[column].isnull().sum() > 0:
            df[column].fillna(df[column].mean(), inplace=True)

    # Impute missing values for categorical columns with the mode
    categorical_columns = ['gender', 'smoking_history']
    for column in categorical_columns:
        if df[column].isnull().sum() > 0:
            df[column].fillna(df[column].mode()[0], inplace=True)

    # Verify missing values are handled
    print("\nMissing values after handling:")
    print(df.isnull().sum())

    print("\nMissing value imputation successful.")

except Exception as e:
    print(f"An error occurred during missing value handling: {e}")

finally:
    print("\nMissing value handling process attempted.")
```

* **Reason for Using:**

* It avoids introducing bias and ensures continuity in numerical datasets.

**Mode Imputation for Categorical Columns:**

**Technique Used:** Missing values in categorical columns (gender, smoking_history) were imputed using the mode of the respective columns (most frequently occurring value).

**Justification for Choices:**

### 2. Handling Outliers

```python
# Handling Outliers & Outlier treatments

try:
    # Importing necessary libraries
    import numpy as np
    import seaborn as sns
    import matplotlib.pyplot as plt

    # Define a function to detect and treat outliers using the IQR method
    def handle_outliers(df, column):
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Visualizing outliers before treatment
        sns.boxplot(x=df[column])
        plt.title(f"Outliers in {column} (Before Treatment)")
        plt.show()

        # Treat outliers by capping them at lower and upper bounds
        df[column] = np.where(df[column] < lower_bound, lower_bound,
                              np.where(df[column] > upper_bound, upper_bound, df[column]))

        # Visualizing outliers after treatment
        sns.boxplot(x=df[column])
        plt.title(f"Outliers in {column} (After Treatment)")
        plt.show()

        return df

    # List of numerical columns to check for outliers
    numerical_columns = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level']

    # Apply outlier treatment to each numerical column
    for column in numerical_columns:
        print(f"Handling outliers for {column}...")
        df = handle_outliers(df, column)

    print("\nOutlier treatment completed successfully.")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    print("\nOutlier detection and treatment attempted.")
```

##### What all outlier treatment techniques have you used and why did you use those techniques?

**What all outlier treatment techniques have you used?**

* **IQR (Interquartile Range) Method:** Outliers were detected using the interquartile range, which is calculated as Q3 - Q1. Values lying outside the range [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR] were identified as outliers.

* **Capping:** The identified outliers were treated by capping them to the nearest boundary within the calculated range.

**Why did you use those techniques?**

* **Preserving the Dataset’s Distribution:** The IQR method is a robust technique that adjusts extreme values without entirely removing them, thus preserving the original distribution and maintaining data integrity.

* **Minimizing Bias:** By capping rather than removing outliers, the risk of introducing bias or losing valuable information was minimized.

* **Suitability for Numerical Data:** The IQR method is particularly effective for numerical features like age, bmi, HbA1c_level, and blood_glucose_level that tend to have natural variability but may also include extreme values.

* **Visualization Insight:** The use of boxplots before and after treatment provided a clear visual understanding of the extent of outliers and the effectiveness of the treatment.

```python
# Manipulate Features to minimize feature correlation and create new features

try:

    # Step 1: Calculate the correlation matrix to identify correlations between features
    correlation_matrix = df.corr()

    # Step 2: Identify and remove highly correlated features (correlation > 0.9)
    highly_correlated_features = set()
    threshold = 0.9  # Define the threshold for high correlation

    # Iterate through the correlation matrix to find features with correlation > 0.9
    for i in range(len(correlation_matrix.columns)):
        for j in range(i):
            if abs(correlation_matrix.iloc[i, j]) > threshold:
                colname = correlation_matrix.columns[i]
                highly_correlated_features.add(colname)

    # Drop the highly correlated features
    df.drop(columns=highly_correlated_features, inplace=True)

    # Step 3: Create new features based on feature interactions or transformations
    # Example: Creating interaction features between BMI and age
    df['bmi_age_interaction'] = df['bmi'] * df['age']

    # Example: Create a chronic condition feature (interaction of hypertension and heart disease)
    df['chronic_condition'] = df['hypertension'] & df['heart_disease']  # True if both are 1, else False

    # Example: Defining custom bin edges for the scaled values
    df['age_group'] = pd.cut(df['age'], bins=[-float('inf'), -1, 0, 1, float('inf')],
                             labels=['Very Young', 'Young', 'Middle-aged', 'Old'])

    # Ensure age_group column is of categorical type
    df['age_group'] = df['age_group'].astype(str)

    # Label encoding for the 'age_group' column
    label_encoder = LabelEncoder()
    df['age_group'] = label_encoder.fit_transform(df['age_group'])

    # Print the updated dataframe for verification
    print("Updated DataFrame with new features and removed highly correlated features:")
    print(df.head())

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    print("Feature manipulation and Create New Feature are completed.")
```

#### 2. Feature Selection

```python
# Select your features wisely to avoid overfitting

try:
    from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif, VarianceThreshold
    from sklearn.ensemble import RandomForestClassifier
    import pandas as pd

    X = df.drop(columns=['diabetes'])  # Features (exclude target column)
    y = df['diabetes']  # Target variable

    # Step 1: Remove low-variance features
    variance_threshold = VarianceThreshold(threshold=0.01)  # Adjust the threshold as needed
    X_var = variance_threshold.fit_transform(X)
    selected_mask_var = variance_threshold.get_support()
    X_var = pd.DataFrame(X_var, columns=X.columns[selected_mask_var])  # Reconstruct DataFrame with selected columns
    print("Step 1: Removed low-variance features.")

    # Step 2: Use statistical tests for feature relevance
    k_best = SelectKBest(score_func=f_classif, k=10)  # Select top 10 features based on ANOVA F-statistic
    X_kbest = k_best.fit_transform(X_var, y)
    selected_features_anova = X_var.columns[k_best.get_support()]
    print(f"Step 2: Selected top features using ANOVA F-test: {selected_features_anova.tolist()}")

    # Step 3: Use mutual information for non-linear relationships
    mutual_info = SelectKBest(score_func=mutual_info_classif, k=10)
    X_mutual = mutual_info.fit_transform(X_var, y)
    selected_features_mutual = X_var.columns[mutual_info.get_support()]
    print(f"Step 3: Selected top features using Mutual Information: {selected_features_mutual.tolist()}")

    # Step 4: Use a feature importance method (e.g., Random Forest)
    model = RandomForestClassifier(random_state=42)
    model.fit(X_var, y)
    feature_importances = pd.Series(model.feature_importances_, index=X_var.columns).sort_values(ascending=False)
    selected_features_rf = feature_importances.head(10).index.tolist()  # Top 10 features
    print(f"Step 4: Selected top features using Random Forest: {selected_features_rf}")

    # Combine all selected features
    final_selected_features = set(selected_features_anova) | set(selected_features_mutual) | set(selected_features_rf)
    print(f"Final selected features: {list(final_selected_features)}")

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    print("Feature selection process completed.")
```

**2. Improving Model Performance**

* Some machine learning algorithms (especially those based on distance metrics like KNN, SVM) can suffer from the curse of dimensionality. As the number of features increases, the model might struggle to capture meaningful patterns, as the feature space becomes sparse.

* Dimensionality reduction techniques, such as PCA, help improve the performance of these models by focusing on the most informative features.

**3. Decreasing Computation Time**

* When you have many features, the training and prediction times of models increase because the algorithms need to process and compute over all the features.

* Reducing the number of features through techniques like PCA can decrease computation time, especially when you're working with large datasets.

**5. Handling Multicollinearity**

* When features are highly correlated with each other, the model may become unstable. This is known as multicollinearity.

* Dimensionality reduction methods like PCA help to transform the features into uncorrelated components, reducing the impact of multicollinearity.

**When Dimensionality Reduction Might Not Be Needed:**

* If the dataset already has a small number of features, applying dimensionality reduction may not be necessary and could even result in loss of information.

* For some machine learning algorithms, such as decision trees or random forests, dimensionality reduction may not be required as they can handle high-dimensional spaces effectively.

**Conclusion:**

Dimensionality reduction is typically needed when you have high-dimensional data that could lead to overfitting, computational inefficiencies, or challenges in interpretation. It is especially useful in improving model performance and reducing overfitting, while also making the data more manageable.

* **Evaluation Metrics:** Common evaluation metrics like accuracy can be misleading. A model that always predicts the majority class (no diabetes) can still have a high accuracy (around 91.5%), but it wouldn't be useful in predicting actual diabetes cases.

Handling imbalanced datasets is crucial because:

* We want to improve the model's ability to detect the minority class (diabetes = 1), which is the main objective of the task.

* **Class Imbalance:** The dataset has an imbalanced distribution between the classes (diabetes vs. no diabetes), where one class (likely "no diabetes") is much more frequent than the other.

* **Synthetic Data Generation:** SMOTE is an over-sampling technique that generates synthetic examples for the minority class by interpolating between existing minority class instances. This helps in balancing the dataset without simply duplicating the minority class instances, which could cause overfitting.

* **Prevents Overfitting:** Unlike random oversampling, which might replicate the minority class instances, SMOTE creates new, synthetic instances. This allows the model to generalize better by providing more varied data for the minority class.

#### 3. Explain each evaluation metric's indication towards business and the business impact of the ML model used.

**Explanation of Evaluation Metrics and Their Business Impact**

**Accuracy:**

* **Definition:** The ratio of correctly predicted instances to the total number of instances.

Accuracy = (True Positives + True Negatives) / Total Instances

* **Business Indication:** Accuracy indicates the overall effectiveness of the model in making correct predictions. For example, an accuracy of 96% means the model correctly classifies 96% of the instances.

* **Business Impact:** A high accuracy rate ensures that the model performs well overall. However, in imbalanced datasets or sensitive cases, accuracy alone may not capture the full picture. For example, focusing on both precision and recall for the minority class (e.g., customers likely to have diabetes) might be more critical than achieving high accuracy alone.

**Precision:**

* **Definition:** The proportion of correctly predicted positive instances out of all instances predicted as positive.

Precision = True Positives / (True Positives + False Positives)

* **Business Indication:** Precision reflects how reliable the positive predictions are. For example, a precision of 87% for Class 1 (diabetes positive) means that 87% of the customers predicted to have diabetes truly have it.

* **Business Impact:** High precision minimizes false positives, which is essential for reducing unnecessary costs or efforts. For example:
* Reducing costs on unnecessary medical tests.
* Avoiding targeting the wrong customers with specialized marketing campaigns.

**Recall:**

* **Definition:** The proportion of actual positive instances correctly identified by the model.

Recall = True Positives / (True Positives + False Negatives)

* **Business Indication:** Recall indicates how well the model captures all relevant cases. For instance, a recall of 60% for Class 1 means the model identifies 60% of individuals with diabetes while missing 40%.

* **Business Impact:** High recall ensures that the model catches most of the important cases:

* In healthcare, missing positive cases (false negatives) can lead to untreated conditions, which might be life-threatening.

* In marketing, missing potential customers could result in lost revenue opportunities.

**F1-Score:**

* **Definition:** The harmonic mean of precision and recall, balancing both metrics.

F1-Score = 2 × ((Precision × Recall) / (Precision + Recall))

* **Business Indication:** F1-score combines precision and recall into a single measure, especially useful when there’s an imbalance between classes. For Class 1, an F1-score of 0.71 shows a balanced performance between precision and recall.

* **Business Impact:** F1-score is critical when both false positives and false negatives are costly:

* In medical diagnostics, balancing both ensures accurate detection while minimizing unnecessary tests.

* In fraud detection, it reduces the risk of missing fraud cases (recall) and wrongly accusing legitimate cases (precision).

**Macro Average:**

* **Definition:** The average of precision, recall, and F1-score across all classes, treating all classes equally.

* **Business Indication:** Macro averages reflect overall model performance across all classes without being biased by the majority class.

* **Business Impact:** Ensures the model doesn’t overlook minority classes, which can be critical in imbalanced datasets. For example:

* Ensuring adequate attention to customers with rare conditions or behaviors.

**Weighted Average:**

* **Definition:** The average of precision, recall, and F1-score across all classes, weighted by class support.

* **Business Indication:** Reflects overall performance but gives more weight to the majority class.

* **Business Impact:** Helps businesses understand real-world performance since it aligns with the dataset distribution.

```python
# ML Model - 3 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, accuracy_score

try:
    print("Initializing hyperparameter tuning for SVC...")

    # Define the model (SVC)
    model = SVC(random_state=42)

    # Define a smaller hyperparameter grid for faster tuning
    param_grid = {
        'C': [0.1, 1, 10],  # Test a few values for C
        'kernel': ['linear', 'rbf']  # Test linear and rbf kernels
    }

    # Use a smaller portion of the training data for faster fitting
    X_train_subsample = X_train.sample(frac=0.2, random_state=42)  # Use 20% of the data
    y_train_subsample = y_train[X_train_subsample.index]

    # Initialize RandomizedSearchCV with reduced iterations and folds for faster execution
    tuner = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_grid,
        n_iter=5,  # Limit iterations to 5 for faster execution
        cv=3,      # 3-fold cross-validation for quicker evaluation
        scoring='accuracy',
        random_state=42,
        verbose=1,
        n_jobs=-1  # Use all processors for parallel computation
    )

    # Fit the model on the subsample
    tuner.fit(X_train_subsample, y_train_subsample)
    print("Hyperparameter tuning completed.")
    print(f"Best Parameters: {tuner.best_params_}")

    # Evaluate the model with the best hyperparameters on the full dataset
    best_model = tuner.best_estimator_
    y_pred = best_model.predict(X_test)

    # Performance metrics
    print("Predictions successfully made on the test set.")
    print(f"Accuracy of the tuned SVC model: {accuracy_score(y_test, y_pred):.2f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    print("Hyperparameter tuning and evaluation process completed.")
```

##### Which hyperparameter optimization technique have you used and why?

I used **RandomizedSearchCV** for hyperparameter optimization. Here's why this technique was chosen:

**Why RandomizedSearchCV?**

**Faster than Grid Search:**

* **GridSearchCV** tests all possible combinations of parameters in the grid, which can be very time-consuming, especially with large hyperparameter spaces.

* **RandomizedSearchCV**, on the other hand, randomly selects combinations from the specified parameter grid, which allows it to explore the search space more efficiently and quickly. By using fewer iterations (n_iter=5), we reduce the computational cost without compromising too much on the model's performance.

**Flexibility:**

* It allows us to focus on the most important hyperparameters, such as C and kernel in an SVC model, and avoid testing excessive combinations.

**Resource-Efficient:**

* It works well in situations where computational resources are limited because we can control the number of iterations and reduce the overall time for hyperparameter tuning.

**Summary:**

**RandomizedSearchCV** was used because it is a more computationally efficient technique than Grid Search, especially with large datasets and hyperparameter grids. It provides a balance between optimizing model performance and saving time and resources.

### 3. Explain the model which you have used and the feature importance using any model explainability tool?

**Random Forest Classifier Explanation:**

The Random Forest Classifier is an ensemble learning method that builds multiple decision trees during training. It works as follows:

**Training:**

* Random Forest creates multiple decision trees by randomly sampling subsets of the data and features. Each tree learns to make decisions based on different patterns in the data.

* The key advantage of using multiple trees is that it reduces overfitting, as each tree is trained independently, and the overall model's prediction is an aggregation (like majority voting) of all the trees' predictions.

**Prediction:**

* When making predictions, each individual tree in the forest gives a prediction, and the Random Forest algorithm takes the majority vote (for classification) or average (for regression) from all the trees to make the final prediction.

**Advantages:**

* **Robustness:** It reduces overfitting, especially when dealing with complex datasets.

* **Handles Imbalanced Data:** It can handle imbalanced datasets by focusing on harder-to-predict cases.

* **Feature Selection:** Random Forest can automatically assess the importance of different features and discard less useful ones.

**Model Explainability:**

We can use model explainability tools like SHAP and Permutation Importance to understand the contribution of each feature to the model's predictions.

**1. SHAP (SHapley Additive exPlanations):**

* SHAP values explain the impact of each feature on the individual predictions. It is based on game theory and calculates the average contribution of a feature to the prediction by comparing all possible combinations of features.

* Why SHAP?: SHAP is a globally consistent and locally interpretable method that provides a clear understanding of how features influence the output.

**How SHAP helps:**

* By calculating SHAP values, we can understand how each feature like bmi, age, or blood_glucose_level impacts the decision-making of the model.

* Features with higher SHAP values have more influence on the model’s predictions. Positive SHAP values indicate that the feature contributes to predicting class 1 (diabetic), while negative values indicate a contribution towards class 0 (non-diabetic).

**Example of SHAP output:**

* **bmi** and **age** are likely to have positive SHAP values, indicating they are highly influential in predicting diabetes.

* **blood_glucose_level** may also show high positive SHAP values, highlighting its importance in determining whether an individual has diabetes.

**2. Permutation Importance:**

* Permutation Importance works by randomly shuffling the values of a feature and measuring the drop in model performance. The greater the decrease in performance, the more important the feature is.

**How Permutation Importance helps:**

* It helps in determining the importance of each feature by evaluating how much each one contributes to the model’s accuracy.

* Example: If bmi is shuffled and the model’s accuracy drops significantly, it indicates that bmi is an important feature in making predictions.

**Steps:**

* Train the Random Forest model.
* Shuffle a feature (e.g., age) and measure the change in accuracy.
* Repeat the process for all features to assess their importance.

**Example Output:**

* **bmi** and **blood_glucose_level** might have the highest importance scores, meaning they are the most influential in predicting diabetes.

* **hypertension** and **heart_disease** could show moderate importance.
